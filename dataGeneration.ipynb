### Extracting frames from video clip and saving as .jpg

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap= cv2.VideoCapture('/Users/aleksi/Desktop/UAL/coding2/coding2_finalproject_data/blackswanvideo/shortswan.mov')
i=0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    cv2.imwrite('/Users/aleksi/Desktop/UAL/coding2/coding2_finalproject_data/blackswanframes2/frame'+str(i)+'.jpg',frame)
    i+=1
 
cap.release()
cv2.destroyAllWindows()
```

### Code for mirroring images to increase input data set size

```python
from PIL import Image
import os
```

```python
def mirror(path, saveto):
    im = Image.open(path)
    flipped_im = im.transpose(Image.FLIP_LEFT_RIGHT)
    flipped_im.save(saveto)

directory = os.fsencode("path_to_folder_containing_images")

for i in os.listdir(directory):
    file = os.fsdecode(i)
    print(file)
    directory2file = "path_to_folder_containing_images" + str(file)
    if file.endswith(".jpg"): 
        mirror(directory2file, "path_to_where_to_save_mirrored_images/"+str(i)+".jpg")
```
